# Handling Duplicate values

```python
df.duplicated().sum()
```

```python
duplicates = df[df.duplicated()]
duplicates.head()

# Here we can see that not all duplicates are not actual duplicates because it’s a transactional structure
# In datasets like Online Retail Dataset, this pattern is common:
# Same InvoiceNo
# Same CustomerID
# Same Country
# BUT different StockCode, Description
# One invoice = multiple products purchased in a single transaction
```

```python
duplicate_rows = df[df.duplicated(keep=False)]
duplicate_rows.head()
```

```python
df_unique = df.drop_duplicates()
```

# Handling Missing Values

```python
# Quantify Missingness Per Column
missing_percent = df_unique.isnull().mean() * 100
missing_percent.sort_values(ascending=False)
```

```python
df_unique['Description'] = df_unique['Description'].fillna('Unknown')
```

```python
df_unique['CustomerID'].isnull().sum()
```

```python
df_unique_no_miss = df_unique[df_unique['CustomerID'].notna()]
df_unique_no_miss['CustomerID'] = df_unique_no_miss['CustomerID'].astype(int).astype(str)
```

```python
# Verify Final Result
df_unique_no_miss['CustomerID'].isnull().sum()
```

# fixing datatypes

```python
df_unique_no_miss.info()
```

```python
df_unique_no_miss['InvoiceDate'] = pd.to_datetime(df_unique_no_miss['InvoiceDate'])
```
